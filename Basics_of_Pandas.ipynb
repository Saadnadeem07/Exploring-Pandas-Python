# Different ways of Creating Data Frames

## By Creating a Dictionary

```python
# Creating a Dictionary for DataFrame
weather_data = {
    'date' : ['01/10/2024','02/10/2024','03/10/2024'],
    'temp' : [32,37,40],
    'windspeed' : [2,8,6],
    'event' : ['Rain','Sunny','Snow']
}

df = pd.DataFrame(weather_data)
df
```

## Creating Tuples for Data Frames

```python
#Creating Tuples for Data Frames
new_data = [
    ('11/09/2024','Monday'), # these are the rows
    ('12/09/2024','Tuesday'),
    ('13/09/2024','Wednesday')
]

df = pd.DataFrame(new_data, columns = ['Date','Day'])
df
```

## Using Dictionary Lists

```python
#By Creating Dictionary Lists
std_data = [
    {'Name': 'Saad Nadeem','Roll No': 9104,'Batch': 21},
    {'Name': 'Saad Habib','Roll No': 9079,'Batch':21},
    {'Name': 'Abdul Basit','Roll No': 9127 ,'Batch': 21}
]

df = pd.DataFrame(std_data)
df
```
